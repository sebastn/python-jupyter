```python
'''This code creates an address book 


***Features***

First name 
Second name
Mail (with validity check)
Phone number (with country code option)
Notes
Edit & Read mode


Optional modes:

--> Check if the name already exists in add mode
--> Edit & Read mode
--> Structure code better
--> Search functionality
--> Better validity check for mail & number
--> GUI ;)

'''

def main():
    
    editVar = False # For edit mode
    showVar = False  # For show mode
    tryAgain = True # For wrong user input
    tryAgain2 = True # For wrong user input
    showUp = False # Helps to print welcome statement just once
    contactId = 0 # Initial 
    
    addressBook = {}
    
    print("Hey user, I am your address book! What is your name? ", '\n')
    userName = input()
    
    while tryAgain == True:
        
        if showUp == False:
            print(f"Hey {userName}! Do you want to edit your address book -- Type: Y for yes or N for no", "\n")
        
        userInput1 = input()
        
        if userInput1.lower() == "y" or userInput1.lower() == "yes":
            editVar = True
            tryAgain = False

        elif userInput1.lower() == "n" or userInput1.lower() == "no":
            print(f"Ok, {userName}. But do you want to show your address book? Again -- Type: Y for yes or N for no", "\n")
            userInput2 = input()
            
            if userInput2.lower() == "y" or userInput2.lower() == "yes":
                print(f"Here is the book!: {addressBook}")
                tryAgain = False
                editVar = False
            
            else:
                print(f"Ok, {userName}, see you next time!")
                tryAgain = False
                editVar = False
                  
        else:
            print()
            print(f"Upps, there is something wrong with your input, {userName}! Again, Y for yes or N for no", "\n")
            tryAgain = True
            showUp = True # Helps to print welcome statement just once


    while editVar == True:
        
        contEdit = "default"
        
        print(f"Ok, {userName}! Let's get started!", "\n")
        name = giveName()
        print(f"You added the name: {name}")
        print()
        
        print("Great! Let's continue!")   
        number = giveNumber()
        print(f"You added the number: {number}")
        print()
        
        print("Great! Let's continue!")   
        email = giveMail()
        print(f"You added the email address: {email}")
        print()
        
        print("Great! Let's continue!")  
        note = giveNote()
        print(f"You added the note: {note}")
        
        addressBook[contactId] = [name, number, email, note]
        contactId += 1
        
        print("Do you want to continue adding contacts to your address book? Type y for YES and n for NO!")
        
        while True:
            contEdit = input()
        
            if contEdit.lower() == "y" or contEdit.lower() == "yes":
                editVar = True
                break

            elif contEdit.lower() == "n" or contEdit.lower() == "no":
                print(f"Ok, {userName}! Here is your current address book: {addressBook}")
                editVar = False
                break
                  
            else:
                print()
                print(f"Upps, there is something wrong with your input, {userName}! Again, Y for yes or N for no", "\n")
        
        
    return 


'''Here are the different input functions!!!'''
              
def giveName():
    print("Add the first name of your new contact: ")
    firstName = input()
    print("and now the last name: ")
    lastName = input() 
    return  firstName + " " + lastName 
                  

def giveNumber():
    
    tryAgain = True # For wrong user input
    
    print("Now let's get the phone number!")
    print("But first let's choose the country code: Type GER for Germany, CH for Switzerland or other for another country code")
    
    while tryAgain == True:
        
        countryCode = input()
    
        if countryCode.lower() == "ger":
            print("Good, you chose Germany with country code: 0049")
            countryCode = "0049"
            tryAgain = False
    
        elif countryCode.lower() == "ch":
            print("Good, you chose Switzerland with country code: 0041")
            countryCode = "0041"
            tryAgain = False
    
        elif countryCode.lower() == "other":
            print("Ok, you want to setup a unique country code: You can type your country code now, e.g. 0056 for Chile...")
            countryCode = input()
            tryAgain = False
    
        else:
            print()
            print(f"Upps, there is something wrong with your country code! Again: Type GER for Germany, CH for Switzerland or other for another country code", "\n")
            tryAgain = True
    
    print("Now, you can type the rest of the phone number")
    number = input()
    number = countryCode + number
        
    return number
        

def giveMail():
    
    tryAgain = True # For wrong user input
    print("Now let's get the e-mail address!")
    print("Type the E-mail address now: ")
    email = input()
    
    while "@" not in email:
        print("That is not a valid e-mail address - the @ is missing...Try again!")
        email = input()
    
    return email
    
    
def giveNote():
    
    tryAgain = True
    print("Now is your chance to edit some additional notes!")
    print("Do you want to add some additional notes? Type y for YES or n for NO! ")
    
    
    while tryAgain:
        noteConfirm = input()
        if noteConfirm.lower() == "y" or noteConfirm.lower() == "yes":
            print("Write your note: ")
            note = input()
            tryAgain = False
    
        elif noteConfirm.lower() == "n" or noteConfirm.lower() == "no":
            print("ok, there will be no notes!")
            countryCode = "0041"
            note = None
            tryAgain = False
    
        else:
            print()
            print(f"Upps, there is something off with your input. Type y for YES or n for NO!", "\n")
            tryAgain = True
                  
    return note
    
```
